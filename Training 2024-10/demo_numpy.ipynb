```python
n = 1000
x = np.random.uniform(1, 10, n)
y = 3 * x + 7 + np.random.randn(n) * 4
```

```python
plt.scatter(x, y, s=8, color='orange')
plt.grid()
plt.show()
```

```python
plt.boxplot(data)
plt.violinplot(data)
plt.grid()
plt.show()
```

```python
import scipy

scipy.__version__
```

```python
slope, intercept, *_ = scipy.stats.linregress(x, y)
```

```python
slope
```

```python
intercept
```

```python
f = lambda x: slope * x + intercept

plt.scatter(x, y, s=8, color='orange')
plt.plot([1, 10], [f(1), f(10)], color='green')
plt.grid()
plt.show()
```
