```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = 'data_paul.csv'
```

```python
df = pd.read_csv(filename)
df
```

```python
df.info()
```

```python
df['PLECS'] = df['PLECS'].replace('N.P.', np.nan)
```

```python
df['LTSpice'] = df['LTSpice'].str.strip(' W').astype('float')
df['Matlab'] = df['Matlab'].str.strip(' W').astype('float')
df['PLECS'] = df['PLECS'].str.strip(' W').astype('float')
```

```python
df['Device'] = df['Device'].astype('category')
```

```python
df.info()
```

```python
df
```

```python
df['P average'] = df[['LTSpice','Matlab','PLECS']].mean(axis=1)
```

```python
df
```

```python
df.columns[:5]
```

```python
agg_data = pd.pivot_table(df,
                          index=list(df.columns[:5]), 
                          columns='Device', 
                          values='P average', 
                          aggfunc='mean').sort_index()

agg_data
```

```python
agg_data.loc[(50,10,3,500000.0),:].transpose().plot(kind='bar', 
                                                    title='Average P')
plt.show()
```
